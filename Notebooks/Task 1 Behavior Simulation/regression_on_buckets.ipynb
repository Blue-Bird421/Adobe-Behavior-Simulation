## COMPANY

```python
df_bucket = pd.read_csv("/kaggle/input/buckets-extra/company_bucket.csv")
```

```python
df = pd.read_csv("/kaggle/input/final-task1/company_task1_df.csv")
df["bucket"] = df_bucket["bucket"]
```

```python
import numpy as np 
import pandas as pd 
import os

from sklearn.model_selection import train_test_split
from xgboost import XGBRegressor
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import LabelEncoder
from catboost import CatBoostRegressor
import warnings

# Suppress all warnings
warnings.filterwarnings("ignore")
```

```python
b0 = df[df["bucket"] == 0]

model = XGBRegressor()
model.load_model("/kaggle/input/catboost/outputs/model_bucket_0.joblib")

dummy_feature1 = [0]*b0.shape[0]
b0["d1"] = dummy_feature1
b0["d2"] = dummy_feature1

embedding_columns = [f'embedding_{i}' for i in range(768)]
features =  ["d1", "d2"] + embedding_columns
X = b0[features]

y_pred = model.predict(X)

b0_final = pd.DataFrame()
b0_final["id"] = b0["id"]
b0_final["pred"] = y_pred
```

```python
b1 = df[df["bucket"] == 1]

model = XGBRegressor()
model.load_model("/kaggle/input/catboost/outputs/model_bucket_1.joblib")

dummy_feature1 = [0]*b1.shape[0]
b1["d1"] = dummy_feature1
b1["d2"] = dummy_feature1

embedding_columns = [f'embedding_{i}' for i in range(768)]
features =  ["d1", "d2"] + embedding_columns
X = b1[features]

y_pred = model.predict(X)

b1_final = pd.DataFrame()
b1_final["id"] = b1["id"]
b1_final["pred"] = y_pred
```

```python
b2 = df[df["bucket"] == 2]

model = CatBoostRegressor()
model.load_model("/kaggle/input/ctboost/outputs/model_bucket_2.catboost")

dummy_feature1 = [0]*b2.shape[0]
b2["log_followers"] = dummy_feature1
b2["labeled_Location"] = dummy_feature1

embedding_columns = [f'embedding_{i}' for i in range(768)]
features =  ["log_followers", "labeled_Location"] + embedding_columns
X = b2[features]

y_pred = model.predict(X)

b2_final = pd.DataFrame()
b2_final["id"] = b2["id"]
b2_final["pred"] = y_pred
```

```python
b4 = df[df["bucket"] == 4]

model = XGBRegressor()
model.load_model("/kaggle/input/catboost/outputs/model_bucket_6.joblib")

dummy_feature1 = [0]*b4.shape[0]
b4["d1"] = dummy_feature1
b4["d2"] = dummy_feature1

embedding_columns = [f'embedding_{i}' for i in range(768)]
features =  ["d1", "d2"] + embedding_columns
X = b4[features]

y_pred = model.predict(X)

b4_final = pd.DataFrame()
b4_final["id"] = b4["id"]
b4_final["pred"] = y_pred
```

```python
final_task1_company = pd.concat([b0_final, b1_final, b2_final, b4_final]).sort_values(["id"])
final_task1_company.to_csv("final_task1_company1.csv")
```
