```python
import os
import pandas as pd
import requests
from PIL import Image 
from io import BytesIO
from tqdm import tqdm
import warnings 
warnings.filterwarnings('ignore')
```

**We have found that some media links in the training and testing data are not working. So we will use black image to replace them.**

```python
black_image_path = os.path.join("/kaggle/working/", "black_image.jpg")
black_image_size = (680, 680)  
black_image = Image.new("RGB", black_image_size, "black")
black_image.save("/kaggle/working/black_image.jpg")
```

```python
df = pd.read_excel("/kaggle/input/adobe-test/behaviour_simulation_test_company.xlsx")
```

```python
df.head()
```

```python
df[['date','time']] = df['date'].str.split(expand = True)
company_counts = df['inferred company'].value_counts()
pattern = r'\[(.*?)\('
df[['media_type']] = df['media'].str.extract(pattern)
pattern = r'\'(.*?)\''
df['url'] = df['media'].str.findall(pattern)
df['num_of_media'] =  df['url'].apply(len)
```

```python
def get_image(image_url):
    response = requests.get(image_url)
    if(response.status_code == 200):
        image = Image.open(BytesIO(response.content)).convert("RGB")
    else:
        image = Image.open("/kaggle/working/black_image.jpg")
    return image
```

```python
# for j in range(10,30):
dir_name = f"/kaggle/working/images_{0}/"
os.makedirs(dir_name, exist_ok=True)
for i in tqdm(range(10000)):
    img = get_image(df["url"][i][0])
    img = img.resize((224,224))
    image_path = dir_name + str(i)+".jpg"
    img.save(image_path)
print(f"{10000} images complete")
```
