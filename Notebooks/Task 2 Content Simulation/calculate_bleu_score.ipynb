```python
import nltk
from nltk.translate.bleu_score import sentence_bleu
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import string

nltk.download('punkt')
```

```python
def clean_and_tokenize(sentence):
    # Convert to lowercase
    sentence = sentence.lower()
    sentence = sentence.replace('\n', '')

    # Remove punctuation
    sentence = sentence.translate(str.maketrans('', '', string.punctuation))

    # Tokenize the sentence
    tokens = nltk.word_tokenize(sentence)

    return tokens

def calculate_bleu_score(reference_sentence, candidate_sentence):
    # Clean and tokenize the input sentences
    reference_tokens = clean_and_tokenize(reference_sentence)
    candidate_tokens = clean_and_tokenize(candidate_sentence)

    # Calculate BLEU score
    bleu_score = sentence_bleu([reference_tokens], candidate_tokens)

    return bleu_score
```
