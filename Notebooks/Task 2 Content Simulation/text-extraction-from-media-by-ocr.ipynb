```python
import pandas as pd
import os
from paddleocr import PaddleOCR
from PIL import Image
import logging
logger = logging.getLogger('PaddleOCR')
logging.basicConfig(level=logging.WARNING)
```

```python
df = pd.read_excel("/kaggle/input/adobe-test/content_simulation_test_company.xlsx")#Path of test dataset for task 2.
df.head()
```

```python
ocr = PaddleOCR(use_angle_cls=True, lang='en')
def perform_ocr_with_paddleocr(image_path):
    
    #ocr = PaddleOCR(use_angle_cls=True, lang='en')  
    result = ocr.ocr(image_path, cls=True)
    
    extracted_texts = []
    for entry in result:
        for item in entry:
            text = item[1][0]  
            extracted_texts.append(text)
    return extracted_texts
```

```python
parent_folder = "/kaggle/input/asobe-test-conetnt-sim-company/content_simulation_test_company_Images" #Path of directory for downloaded media i.e, Image, Video/GIF Thumbnail. 
ids = df['id']
texts = []

for image_id in ids:
    if image_id%1000==0:
        print(f"Processed {image_id}th examples")
        
    image_num = str(image_id-1)+'.jpg'
    image_path = os.path.join(parent_folder, image_num)
    
    try:
        text = perform_ocr_with_paddleocr(image_path)
        texts.append(text)
        
    except:
        text = ["No Text content present on Image"]
        texts.append(text)
    
```

```python
ocr_text = pd.DataFrame()
ocr_text['text_extracted'] = texts
ocr_text.head()
```

```python
ocr_text.to_csv("/kaggle/working/ocr_text_company.csv")
```
